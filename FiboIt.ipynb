```python
from time import time
import matplotlib.pyplot as plt
```

```python
def fib(n):
    start_time = time()
    if(n==0):
        y=0
        #print "fibonacci number 0"
    else:
        x=0
        y=1
        for i in range(1,n):
            z=(x+y)
            x=y
            y=z
        #print "fibonacci number",y
    return time() - start_time
```

```python
n=15000
tiempos = []
numeros = list(range(n+1))

for i in range (n+1):
    tiempos = tiempos + [fib(i)]

plt.plot(numeros, tiempos)
plt.xlabel('n')
plt.ylabel('segundos')
plt.title('Tiempo para hallar fibonacci');
```

```python
plt.scatter(numeros, tiempos)
plt.xlabel('n')
plt.ylabel('segundos')
plt.title('Tiempo para hallar fibonacci');
```
